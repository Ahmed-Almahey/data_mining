```python
x = df[['income','score']].values
```

```python
x
```

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler, RobustScaler
```

```python
scaler = StandardScaler()
```

```python
x_scaled = scaler.fit_transform(x)
```

```python
x_scaled
```

```python
#kmeans clustering and modeling 
from sklearn.cluster import KMeans
km = KMeans(n_clusters = 5, init='k-means++', random_state = 0)
```

```python
km.fit(x_scaled)
```

```python
y_pred = km.predict(x_scaled)
```

```python
y_pred
```

```python
x_df = pd.DataFrame(data = x)
```

```python
km.labels_
```

```python
x_df['Clusters']= y_pred
```

```python
x_df
```

```python
x[y_pred==0].shape
```

```python
plt.figure(figsize=(6,6))
plt.scatter(x[y_pred==0,0], x[y_pred==0,1], s =100, c= 'red',label='cluster 1') 
plt.scatter(x[y_pred==1,0], x[y_pred==1,1], s =100, c= 'gray',label='cluster 2') 
plt.scatter(x[y_pred==2,0], x[y_pred==2,1], s =100, c= 'green',label='cluster 3') 
plt.scatter(x[y_pred==3,0], x[y_pred==3,1], s =100, c= 'blue',label='cluster 4') 
plt.scatter(x[y_pred==4,0], x[y_pred==4,1], s =100, c= '#6F4F37',label='cluster 5') 
plt.title("Cluster of Customers")
plt.xlabel("Annual Income (k$)")
plt.ylabel("Spending Score")
plt.legend()
```

```python
km.inertia_
```

```python
inertia = []
k_range = range(2,11)
for i in k_range:
    kmean = KMeans(n_clusters=i,init='k-means++',random_state=0)
    kmean.fit(x_scaled)
    inertia.append(kmean.inertia_)

fig, ax = plt.subplots(figsize=(12,8)) 
sns.lineplot(x=k_range,y=inertia)
plt.title("The Elbow Method")
plt.xlabel("No of Clusters")
plt.ylabel("Inertia")
plt.show()
```
